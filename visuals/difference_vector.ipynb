```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ideal = np.array([1, 0])
vec = np.array([0.81, 0.42])
comp = ideal - vec

H = np.array([ideal, 
              vec,
              comp])

origin = np.array([[0, 0, vec[0]],
                   [0, 0, vec[1]]]) # origin point

plt.quiver(*origin, H[:,0], H[:,1], 
           color=['tab:green','tab:blue', 'tab:red'], 
           angles='xy', scale_units='xy', scale=1,
           width=0.007)


# Define legend elements
legend_colors = ['tab:green', 'tab:blue', 'tab:red']
legend_labels = ['Ideal Vector', 'General Vector', 'Difference Vector']

# Optional: Create empty artist objects (lines in this case)
legend_handles = [plt.Line2D([], [], color=c, marker='', linestyle='-') for c in legend_colors]

# Create the legend
plt.legend(legend_handles, legend_labels, title="Legend")

alpha = 0.4
plt.axhline(y=0, color='k', alpha=alpha)
plt.axvline(x=0, color='k', alpha=alpha)
plt.axis('equal')
plt.xlabel("Cosine Similarity")
plt.ylabel("Relative Semantic Difference")
plt.title("Distance From Perfect Similarity")
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1)
plt.grid(visible=True, which='major')
plt.show()
```
